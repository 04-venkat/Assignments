```python
x = np.arange(1, 11)
y = 3*x+5
plt.title("title")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x,y)
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("sine")
plt.plot(x, y, color='r')
plt.show()
```

```python
x = np.arange(0, 10)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 10)
y = np.arange(0, 10)
plt.subplot(2, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(2, 2, 2)
plt.plot(x, y, 'g--')
plt.subplot(2, 2, 3)
plt.plot(x, y, 'b--')
plt.subplot(2, 2, 4)
plt.plot(x, y, 'b')
plt.show()
```
