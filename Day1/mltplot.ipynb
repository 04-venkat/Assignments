```python
import numpy
import matplotlib.pyplot as plt
```

```python
#Question 1

a = numpy.arange(40,50)
b = numpy.arange(50,60)
plt.title("Simple Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.plot(a, b)
plt.show()
```

```python
#Question 2

days = numpy.arange(1, 8) 
sales1 = [160, 150, 140, 145, 175, 165, 180] 
sales2 = [70, 90, 160, 150, 140, 145, 175] 
# plt.figure(figsize = (6, 5), dpi = 150)
plt.title("Days v. Sales")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.plot(days, sales1)
plt.plot(days, sales2)
plt.show()
```

```python
#Question 3

x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3, 3, 1)
plt.plot(x, y1, 'r')
plt.subplot(3, 3, 2)
plt.plot(x, y2, 'r--')
plt.subplot(3, 3, 3)
plt.plot(x, y3, 'b')
plt.subplot(3, 3, 4)
plt.plot(x, y4, 'b--')
plt.subplot(3, 3, 5)
plt.plot(x, y5)
plt.show()
```
